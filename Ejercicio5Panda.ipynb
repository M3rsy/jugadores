**Crear un nuevo dataframe llamado delanteros y realice el filtro de todos los delanteros (Forward) de todos los equipos. **

```python
import pandas as pd
import numpy as np
delanteros = pd.read_csv ('https://raw.githubusercontent.com/M3rsy/jugadores/main/jugadores.csv', index_col=0)
#Covid = pd.read_csv ('https://raw.githubusercontent.com/owid/covid-19-data/master/public/data/owid-covid-data.csv', index_col=0)
delanteros.loc[delanteros['position'] == "Forward"[0:10]]



```

**Seleccione los jugadores con mas de 10 goles y solo las columnas ( team, player_name, position, todas las columnas que inicien con goals)**  

```python
print(delanteros[delanteros["goals"]>10]['player_name'].sort_values())
```

```python
print(delanteros[['goals', 'player_id', 'position']])
```

**Los jugadores que han anotado goles con la cabeza (goals_headed) y que so defensas (position= Defender**

```python
delanteros.sort_values(['goals_headed', 'position'], ascending=True)
```

**Guardar en un dataframe los jugadores que interceptions > 40 y recoveries > 100**

```python
print(delanteros[delanteros["interceptions"]>40]['player_name'].sort_values())


```

```python
print(delanteros[delanteros["recoveries"]>100]['player_name'].sort_values())
```

```python
delanteros = pd.DataFrame(delanteros)
delanteros.to_csv('jugadores.csv', header=False, index=False)
delanteros
```
